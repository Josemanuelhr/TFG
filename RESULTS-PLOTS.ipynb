```python
import numpy as np
import matplotlib.pyplot as plt
 
# set width of bar
barWidth = 0.15
fig = plt.subplots(figsize =(20, 8))
plt.ylim(0, 1)
 
# RME, CC, R2
SVM = [0.87, 0.3, 0.2, 0.524]
kNN = [0.93, 0.2, 0.5, 0.4224]
DT = [0.80, 0.4, 0.4, 0.2324]
GP = [0.95, 0.63, 0.12, 0.624]
RF = [0.99, 0.13, 0.762, 0.8]
 
# Set position of bar on X axis
br1 = np.arange(len(SVM))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
br4 = [x + barWidth for x in br3]
br5 = [x + barWidth for x in br4]
 
# Make the plot
plt.bar(br1, SVM, width = barWidth, edgecolor ='grey', label ='SVM')
plt.bar(br2, kNN, width = barWidth, edgecolor ='grey', label ='kNN')
plt.bar(br3, DT, width = barWidth, edgecolor ='grey', label ='DT')
plt.bar(br4, GP, width = barWidth, edgecolor ='grey', label ='GP')
plt.bar(br5, RF, width = barWidth, edgecolor ='grey', label ='RF')
 
# Adding Xticks
plt.xlabel('Metric', fontweight ='bold', fontsize = 18)
plt.ylabel('Score', fontweight ='bold', fontsize = 18)
plt.xticks([r + barWidth*2 for r in range(len(SVM))], ['Accuracy', 'Precision', 'Recall', 'f1'], fontsize=16)
plt.yticks(fontsize=16)

plt.legend(fontsize=16)
plt.savefig("Images-Test/test-barras.pdf", bbox_inches='tight')
plt.show()
```
